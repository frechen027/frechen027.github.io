## 代码示例

下面我们将运行一段简单的 Python 代码，并绘制一个图表。

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成正弦波数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Sample Plot from Notebook')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
